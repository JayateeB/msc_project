```python
import pandas as pd
import os
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
path = "/Users/jay/MSC_WSBDA/MSc_Thesis/Msc_project/Data/"
```

```python


INPUT_PATH = "/Users/jay/MSC_WSBDA/MSc_Thesis/Msc_project/Data/Results/after_time/corrected"

df_list = []

for filename in os.listdir(INPUT_PATH):
    df = pd.read_csv(f"{INPUT_PATH}/{filename}")
    df_list.append(df)

df = pd.concat(df_list)

tick_spacing = 1

columns = ['simulation_start_time', 'precision','recall','f1_score']
df = df[columns]
pf = df
# df = df.set_index('simulation_start_time')
# df.sort_index(inplace=True)
# plt = df.plot.line()
# plt.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
```

```python
with plt.style.context(('ggplot')):
    plt.figure(figsize=(15,9))
    pf = pf.sort_values(by=['simulation_start_time'])
    x = np.array(pf['simulation_start_time'])
    precision = np.array([round(x,2) for x in list(pf['precision'].values)])
    recall = np.array([round(x,2) for x in list(pf['recall'].values)])
    f1_score = np.array([round(x,2) for x in list(pf['f1_score'].values)])
    plt.plot(x,precision,color='green',label='Precision Score')
    plt.plot(x,recall,color='blue',label='Recall Score')
    plt.plot(x,f1_score,color='red',label='F1 Score')
    
    
    plt.xlabel('Simulation Start Time',fontsize=15)
    plt.ylabel('Accuracy Score',fontsize=15)
    plt.legend(loc='upper left',fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title("")
    for a, b in zip(x, precision):
        plt.text(a, b, b, ha='right', va='bottom', fontsize=15,color='green')
    for a, b in zip(x, recall):
        plt.text(a, b, b, ha='right', va='bottom', fontsize=15,color='blue')
    for a, b in zip(x, f1_score):
        plt.text(a, b, b, ha='right', va='bottom', fontsize=15,color='red')
    #plt.savefig(path+"/graph/givenchy_simulation_accuracy_after_current_time.png")
    plt.show()
```
