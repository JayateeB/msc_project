```python
def precision_recall(network_simulation, users, sim_hours=hours):

    simulation_comparison = pd.concat([network_simulation['time_lapsed'],users['time_lapsed']],axis = 1)
    column_names = ['nw_time_lapsed','usr_time_lapsed']
    simulation_comparison.columns = column_names
    
    result = {"hour": [], "precision": [], "recall": [], "f1_score": []}
    for hour in sim_hours:
        df = simulation_comparison[simulation_comparison['usr_time_lapsed'] <= hour]
        df['both_infected'] = df.apply(lambda x: 1 if (np.isnan(x['nw_time_lapsed']) == False)
                                                                                  & (np.isnan(x['usr_time_lapsed']) == False)
                                                                                   & ((x['nw_time_lapsed'] < hour) & 
                                                                                      (x['usr_time_lapsed'] <= hour))
                                                                             else 0,axis = 1)

        true_positive = df['both_infected'].sum()
        total_predicted_positive = len(df[np.isnan(df['nw_time_lapsed']) == False])
        precision = true_positive/total_predicted_positive
        total_actual_positive = len(df[np.isnan(df['usr_time_lapsed']) == False])
        recall = true_positive/total_actual_positive
        f1_score = (2*(precision * recall))/(precision + recall)
        result["hour"].append(hour)
        result["precision"].append(precision)
        result["recall"].append(recall)
        result["f1_score"].append(f1_score)

    return pd.DataFrame.from_dict(result)
        
```

```python
def precision_recall_by_time_lapsed(net_sim, users, sim_hours=hours):
    result = {"hour": [], "precision": [], "recall": [], "f1_score": []}
    
    for hour in sim_hours:
        sim_ids = net_sim[(net_sim["time_lapsed"].isnull() == False) & \
                          (net_sim["time_lapsed"] <= hour)]["id"].tolist()
        
        actual_ids = users[(users["time_lapsed"].isnull() == False) & \
                           (users["time_lapsed"] < hour + 30)]["id"].tolist() 
        
        sim_id_set = set(sim_ids)
        actual_id_set = set(actual_ids)
        true_positive = len(sim_id_set.intersection(actual_id_set))
        total_predicted_positive = len(sim_id_set)
        total_actual_positive = len(actual_id_set)
        #print(f"hour: {hour}, total_predicted_positive: {total_predicted_positive}, total_actual_positive: {total_actual_positive}, true_positive: {true_positive}")
        precision = 0 if total_predicted_positive == 0 else true_positive/total_predicted_positive
        recall = 0 if total_actual_positive == 0 else  true_positive/total_actual_positive
        #print(f"hour: {hour}, precision: {precision}, recall: {recall}")
        f1_score = 0 if (precision == 0 and recall == 0)  else (2*(precision * recall))/(precision + recall)
        result["hour"].append(hour)
        result["precision"].append(precision)
        result["recall"].append(recall)
        result["f1_score"].append(f1_score)
        
    return pd.DataFrame.from_dict(result)     
    
```

```python
def nyc_inaccurate_results(net_sim, users, sim_hours):
    result = {"hour": [], "count": []}
    prev_hour = 420
    for hour in sim_hours:
        sim_ids = net_sim[(net_sim["time_lapsed"].isnull() == False) & \
                          (net_sim["time_lapsed"] >= prev_hour) & \
                          (net_sim["time_lapsed"] < hour)]["id"].tolist()
        
        actual_ids = users[(users["time_lapsed"].isnull() == False) & \
                           (users["time_lapsed"] >= prev_hour) & \
                           (users["time_lapsed"] < hour)]["id"].tolist()   
        wrong_prediction = set(sim_ids) - set(actual_ids)
        result["hour"].append(prev_hour)
        result["count"].append(len(wrong_prediction))
        prev_hour = hour
    
    return pd.DataFrame.from_dict(result)     
        
```
