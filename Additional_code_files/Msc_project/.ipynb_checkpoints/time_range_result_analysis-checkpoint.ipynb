```python
import pandas as pd
import os
import matplotlib.ticker as ticker
```

```python

INPUT_PATH = "/Users/jay/MSC_WSBDA/MSc_Thesis/Msc_project/Data/Results/after_time"

df_list = []

for filename in os.listdir(INPUT_PATH):
    df = pd.read_csv(f"{INPUT_PATH}/{filename}")
    df_list.append(df)

df = pd.concat(df_list)

tick_spacing = 1

columns = ['simulation_start_time', 'f1_score']
df = df[columns]
df = df.set_index('simulation_start_time')
df.sort_index(inplace=True)
plt = df.plot.line()
plt.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

```
